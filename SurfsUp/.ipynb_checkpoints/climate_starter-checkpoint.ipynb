```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement = Base.classes.measurement
station = Base.classes.measurement
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
most_recent_date = session.query(measurement.date).order_by(measurement.date.desc()).first()
print("Most recent date: ", most_recent_date)
```

```python
# Create an inspector for the engine
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for column in columns:
    print(column['name'], column['type'])
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Calculate the date one year from the last date in data set.
query_date = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print("Query Date: ", query_date)
```

```python
# Perform a query to retrieve the data and precipitation scores
annual_prcp = session.query(measurement.date, measurement.prcp).filter(measurement.date >= query_date).all()
annual_prcp
```

```python
# Save the query results as a Pandas DataFrame. Explicitly set the column names
column_names = ['Date', 'Precipitation']
annual_prcp_df = pd.DataFrame(annual_prcp, columns=column_names)
annual_prcp_df
```

```python
# Sort the dataframe by date
# Sort the DataFrame by date in ascending order
ordered_annual_prcp_df = annual_prcp_df.sort_values('Date', ascending=True)

# Drop NaN values in the 'Precipitation' column
ordered_annual_prcp_df = ordered_annual_prcp_df.dropna(how='any')

# Set 'Date' as the index
ordered_annual_prcp_df.set_index('Date', inplace=True)

# Display the resulting DataFrame
ordered_annual_prcp_df
```

```python
plt.figure(figsize=(40, 25))
ordered_annual_prcp_df.plot(y='Precipitation')
plt.xlabel('Date')
plt.ylabel('Precipitation (inches)')
plt.title('Precipitation in Hawaii')
plt.legend(loc='upper right')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
ordered_annual_prcp_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
unique_stations_count = session.query(station.station).distinct().count()

# Print the number of unique stations
print(f'The number of unique stations is: {unique_stations_count}')
```

```python
# Query to count the number of stations
station_count_query = session.query(func.count(station.station))

# Execute the query and retrieve the result
station_count = station_count_query.one()[0]
```

```python
# Create an inspector for the engine
inspector = inspect(engine)
columns = inspector.get_columns('station')
for column in columns:
    print(column['name'], column['type'])
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.
station_activity = session.query(station.station, func.count()) \
    .group_by(station.station) \
    .order_by(func.count().desc()) \
    .all()

# Print the stations and their counts in descending order
for station, count in station_activity:
    print(f"Station: {station}, Count: {count}")
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
# Replace 'MOST_ACTIVE_STATION_ID' with the actual ID of the most active station
most_active_station_id = 'USC00519281'

# Query to calculate the lowest, highest, and average temperature for the most active station
temperature_stats = session.query(
    func.min(measurement.tobs).label('lowest_temperature'),
    func.max(measurement.tobs).label('highest_temperature'),
    func.avg(measurement.tobs).label('average_temperature')
).filter(measurement.station == most_active_station_id).first()

# Print the temperature statistics
print(f"Lowest Temperature: {temperature_stats.lowest_temperature}°F")
print(f"Highest Temperature: {temperature_stats.highest_temperature}°F")
print(f"Average Temperature: {temperature_stats.average_temperature:.2f}°F")
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
most_recent_date = session.query(func.max(measurement.date)) \
    .filter(measurement.station == most_active_station_id) \
    .one()[0]
print(most_recent_date)
```

```python
# Calculate the date 12 months ago from the most recent date
query_date = dt.date(2017, 8, 18) - dt.timedelta(days=365)
print("Query Date: ", query_date)
```

```python
# Query the last 12 months of temperature observation data for the most active station
temperature_data = session.query(measurement.date, measurement.tobs) \
    .filter(measurement.station == most_active_station_id) \
    .filter(measurement.date >= query_date) \
    .all()
```

```python
# Create a DataFrame from the query results
temperature_df = pd.DataFrame(temperature_data, columns=['Date', 'Temperature'])
temperature_df
```

```python
# Create a histogram plot
plt.figure(figsize=(10, 6))
plt.hist(temperature_df['Temperature'], bins=12, edgecolor='black', alpha=0.7)
plt.xlabel('Temperature (°F)')
plt.ylabel('Frequency')
plt.title('Temperature Observation Data for the Last 12 Months')
plt.legend(['tobs'], loc='best')
plt.grid(True)
```

```python
# Close Session
session.close()
```
